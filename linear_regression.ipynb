```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/height-weight.csv')
```

```python
df
```

```python
plt.scatter(df['Height'], df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
x=df['Weight']
y=df['Height']
```

```python

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(np.array(X_train).reshape(-1,1),y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred = model.predict(np.array(X_test).reshape(-1,1))

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"R-squared: {r2}")
print(f"Mean Squared Error: {mse}")
```

```python
plt.scatter(df['Weight'], df['Height'])
plt.plot(df['Weight'], model.predict(np.array(df['Weight']).reshape(-1,1)), color='red')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Height vs Weight with Linear Regression Line')
plt.show()
```

```python
# Enter a new weight value here to get a height prediction
new_weight = 75

predicted_height = model.predict(np.array([[new_weight]]))
print(f"For a weight of {new_weight} kg, the predicted height is: {predicted_height[0]:.2f} cm")
```
